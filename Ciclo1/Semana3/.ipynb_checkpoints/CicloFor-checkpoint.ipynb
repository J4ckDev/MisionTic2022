12. Diseñar una función que permita calcular una aproximación de la función arco tangente para cualquier valor $x ∈ [−1, 1]$, utilizando los primeros n términos de la serie de Maclaurin (al evaluar esta función el resultado que se obtiene está expresado en radianes) $\sum_{i=0}^{n}\frac{(-1)^ix^{2i+1}}{(2i+1)}$



## Solución

### Importación de librerias

```python
from math import factorial
```

## 1. R/

```python
for num in range(101):
    print(f"{num}^2 = {num**2}")
```

## 2. R/

```python
print("Pares\tImpares")
print("\t"+str(1))
for numero in range(2,1001):
    if numero % 2 == 0:
        print(numero, end="\t")
    else:
        print(numero, end="\n")
```

## 3. R/

```python
max_number = int(input("Ingrese un número natural mayor a 2: "))
if max_number >= 2:
    print(f"Los numeros pares entre {max_number} y 2 son:")
    for num in range(max_number,1,-1):
        if num % 2 == 0:
            print(num, end=" ")
else:
    print("Ingresaste un número menor :(")
```

## 4. R/

```python
max_number = int(input("Ingrese un número positivo mayor a 1: "))
previous = 1
if max_number >= 1:
    print(f"El factorial de los numeros entre 1 y {max_number} son:")
    for num in range(1,max_number+1):
        previous *= num
        print(f"{num}! = {previous}")
else:
    print("Ingresaste un número menor :(")
```

## 5. R/

```python
print("Calculadora de la potencia 2^n")
n = int(input("Ingrese el valor de n: "))
resultado = 1
if n != 0:
    for num in range(n):
        resultado *= 2
print(f"2^{n} = {resultado}")
```

## 6. R/

```python
print("Calculadora de potencias x^n")
x = int(input("Ingrese el valor de x: "))
n = int(input("ingrese el valor de n: "))
resultado = 1
if n != 0:
    for num in range(n):
        resultado *= x
print(f"{x}^{n} = {resultado}")
```

## 8. R/

```python
def fun_exp(exponent,n):
    result = 0
    for i in range(n+1):
        result += (exponent**i)/(factorial(i))
    return result

print("Obtener el valor de e^x a partir de la serie de Maclaurin")
exponent = float(input("Ingresa el valor de x: "))
n = int(input("Ingresa el número de terminos de Maclaurin: "))
result = fun_exp(exponent,n)
print(f"e^{exponent} = {result}")
```

## 9. R/

```python
def fun_sin(angle,n):
    result = 0
    for i in range(n+1):
        numerator = ((-1)**i)*(angle**((2*i)+1))
        denominator = factorial((2*i)+1)
        result += numerator/denominator
    return result

print("Obtener el valor de sen(x) a partir de la serie de Maclaurin")
angle = float(input("Ingresa el valor del ángulo x en radianes: "))
n = int(input("Ingresa el número de terminos de Maclaurin: "))
result = fun_sin(angle,n)
print(f"sen({angle}) = {result}")
```

## 10. R/

```python
def fun_cos(angle,n):
    result = 0
    for i in range(n+1):
        numerator = ((-1)**i)*(angle**(2*i))
        denominator = factorial(2*i)
        result += numerator/denominator
    return result

print("Obtener el valor de cos(x) a partir de la serie de Maclaurin")
angle = float(input("Ingresa el valor del ángulo x en radianes: "))
n = int(input("Ingresa el número de terminos de Maclaurin: "))
result = fun_cos(angle,n)
print(f"cos({angle}) = {result}")
```

## 11. R/

```python
def fun_ln(x,n):
    result = 0
    for i in range(n+1):
        first_part = 1/((2*i)+1)
        second_part = (((x**2)-1)/((x**2)+1))**((2*i)+1)
        result += first_part*second_part
    return f"ln({x}) = {result}"

print("Obtener el valor de ln(x) a partir de la serie de Maclaurin")
x = float(input("Ingresa un valor de x mayor a 0: "))
n = int(input("Ingresa el número de terminos de Maclaurin: "))
result = print(fun_ln(x,n)) if x > 0 else print("Número fuera del rango :(")
```

## 12. R/

```python
def fun_arctan(value,n):
    result = 0
    for i in range(n+1):
        numerator = ((-1)**i)*(value**((2*i)+1))
        denominator = (2*i)+1
        result += numerator/denominator
    return f"arctan({value}) = {result}"

print("Obtener el valor de arctan(x) a partir de la serie de Maclaurin")
value = float(input("Ingresa el valor de x entre -1 y 1: "))
n = int(input("Ingresa el número de terminos de Maclaurin: "))
result = print(fun_arctan(value,n)) if (value >= -1 and value <= 1) else print("Número fuera del rango :(")
```
