# Problema

Para el JSON con la siguiente estructura:  
```json
{
    "jadiazcoronado":{
        "nombres": "Juan Antonio",
        "apellidos": "Dı́az Coronado",
        "edad":19,
        "colombiano":true,
        "deportes":["Fútbol","Ajedrez","Gimnasia"]
    },
    ...
    "dmlunasol":{
        "nombres": "Dorotea Maritza",
        "apellidos": "Luna Sol",
        "edad":25,
        "colombiano":false,
        "deportes":["Baloncesto","Ajedrez","Gimnasia"]
    }
}
```

Cree un programa que lea ese archivo JSON y realice lo siguiente:

1. Imprima los nombres completos (nombre y apellidos) de las personas que practican el deporte ingresado por el usuario.
2. Imprima los nombres completos (nombre y apellido) de las personas que estén en un rango de edades dado por el usuario.
3. Cree un JSON de deportes con la siguiente estructura:
```json
{
"Ajedrez":["jadiazcoronado",...,"dmlunasol"],
"Futbol":["jadiazcoronado",...],
"Gimnasia":["jadiazcoronado,...,"dmlunasol"],
...
"Baloncesto":[...,"dmlunasol"]
}
```

4. Desarrolle un programa que lea dos archivos JSON, y encuentre los componentes clave:valor que son iguales en ambos. Genere un nuevo archivo JSON con las coincidencias exactas entre los dos archivos.
5. Desarrolle un programa que lea un archivo JSON, en el cual se encuentran las notas de los estudiantes del curso. Cada llave corresponde al código de cada estudiante, y su valor es una lista con las notas obtenidas en las actividades del curso. Se debe generar un nuevo archivo JSON que para uno de los estudiantes solo guarde el promedio de las notas obtenidas. 
6. Desarrollar un programa que lea un archivo JSON que contiene una serie de cadenas de caracteres en minúscula, cada una con su propia llave. Estas llaves tienen una codificación, a forma de encriptación, en donde las vocales están descritas como otros sı́mbolos: \\$ en vez de a, # en vez de e, * en vez de i, ¬ en vez de o, y + en vez de u. Una vez leı́do el archivo, realice una desencriptación de todas las cadenas, es decir, convierta los sı́mbolos a sus vocales correspondientes (si la cadena de entrada es ”h¬l \\$”, la cadena resultante serı́a ”hola”), y guarde el resultado en un nuevo archivo JSON.


## Solución

### 1 - 3. R/

````python
import json
from pprint import pprint

def show_people_by_sport(people,sport):
    print(f"Personas que practican {sport}")
    for person in people:
        sports = people[person]['deportes']
        if sport in sports:
            print(f" - {people[person]['nombres']} {people[person]['apellidos']}")
            
def show_people_by_age_range(people,min_age,max_age):
    print(f"Personas entre {min_age} a {max_age} años")
    for person in people:
        age = people[person]['edad']
        if age >= min_age and age <= max_age:
            print(f" - {people[person]['nombres']} {people[person]['apellidos']}")
            
def get_all_sports(people):
    all_sports = []
    for person in people:
        sports = people[person]["deportes"]
        for sport in sports:
            if sport not in all_sports:
                all_sports.append(sport)
    return all_sports

def get_usernames_by_sport(people,sport):
    all_people = []
    for person in people:
        sports = people[person]['deportes']
        if sport in sports:
            all_people.append(person)
    return all_people

def create_json_by_sports(people):
    json_data = {}
    sports = get_all_sports(people)
    for sport in sports:
        json_data[sport] = get_usernames_by_sport(people,sport)
    with open("./sports.json", "w") as sports_file:
        json.dump(json_data,sports_file)

with open('./people.json','r') as people_data:
    people = json.load(people_data)
    show_people_by_sport(people,"Natación")
    show_people_by_age_range(people,30,45)
    create_json_by_sports(people)

with open('./sports.json','r') as sports_data:
    pprint(json.load(sports_data))
    
````

### 4. R/

````python
'''
    Compare JSON with the next format
    [
        {
            "key1" : value1,
            "key2" : value2,
            "key3" : value3,
        },
        {
            "key1" : value4,
            "key2" : value5,
            "key3" : value6,
        }
    ]
'''

def compare_two_json(main_json,json_to_compare):
    equal_data = {}
    flag = 0
    for main_data in main_json:        
        for data in json_to_compare:
            for key in data:                
                value = data[key]
                if key in main_data and main_data[key] == value:
                    if equal_data.get(flag) == None:
                        equal_data[flag] = {}
                    else:
                        if equal_data[flag].get(key) != None:
                            flag += 1
                            equal_data[flag] = {}
                            
                    equal_data[flag][key] = value
                
        
    with open('./equalData.json','w') as equal_data_file:
        json.dump(equal_data,equal_data_file)

main_json = []
json_to_compare = []
with open('./users.json','r') as users_data:
    main_json = json.load(users_data)
    
with open('./debtors.json','r') as debtors_data:
    json_to_compare = json.load(debtors_data)
    
compare_two_json(main_json,json_to_compare)
````

### 5. R/

````python
def students_final_grades(grades):
    sum_grades = 0.0
    total_grades = 0
    final_grades = {}
    for student_key in grades:
        student_grades = grades[student_key]
        for grade in student_grades:
            sum_grades += float(grade)
            total_grades += 1
        final_grades[student_key] = str(sum_grades/total_grades)
        sum_grades=0
        total_grades=0
    with open('./finalGrades.json','w') as final_grades_data:
        json.dump(final_grades,final_grades_data)

with open('./grades.json','r') as grades_data:
    grades = json.load(grades_data)
    students_final_grades(grades)
````

### 6. R/

````python
def decrypt(data_encrypted):
    decrypted = {}
    for item in data_encrypted:
        data_encrypted[item] = data_encrypted[item].replace("$","a")
        data_encrypted[item] = data_encrypted[item].replace("#","e")
        data_encrypted[item] = data_encrypted[item].replace("*","i")
        data_encrypted[item] = data_encrypted[item].replace("¬","o")
        data_encrypted[item] = data_encrypted[item].replace("+","u")
        decrypted[item] = data_encrypted[item]
    with open('./decrypted.json','w') as decrypted_file:
        json.dump(decrypted,decrypted_file)
        
with open('./encrypted.json','r') as encrypted_file:
    data_encrypted = json.load(encrypted_file)
    decrypt(data_encrypted)
````
